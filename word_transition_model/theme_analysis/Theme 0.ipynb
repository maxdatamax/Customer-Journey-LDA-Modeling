<h1> Theme 0: Reviewing and Disputing Transactions 

<h2> Transition 1 (Top) </h2> webevent.view.transactions.and.details.success->webstc.dispute 

<h4> What's going on</h4> User looks at previous transactions and then initiates a dispute. This takes place online.
<br><br>

<h2> Transition 2 </h2> webstc.dispute->webevent.view.transactions.and.details.success

<h4> What's going on </h4> This is the top transition but in reverse order. This is likely after the dispute, the user goes back to their transactions.
<br><br>

<h2> What's Interesting</h2> While going from reviewing web transactions to a dispute makes sense, it is an unexpected customer behaviour to want to go back to the transactions after filing the dispute. This makes sense on a psychological level: the user is worried and wants to see if any other fradulent transactions were made.
<br><br>

<h2> What we can do with this data </h2>
If customers are going back to their transactions after filing a dispute then why not make the transactions visible on the dispute screen. This way, customers can review their charges as they file the dispute. This way, if there are multiple charges that the user wants to dispute, the user will be able to include all of them on a single dispute. This is more efficient than the user catching a new fraudulent/erroneous charge and opening another dispute after submitting the first one.
